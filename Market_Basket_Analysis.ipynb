```python
top_products = df["itemDescription"].value_counts()

top_products.head(10)
```

```python
df["Transaction"] = (
    df["Member_number"].astype(str)
    + "_"
    + df["Date"])
```

```python
print(
    "Number of Transactions:",
    df["Transaction"].nunique())
```

```python
basket = (
    df.groupby(
        ["Transaction", "itemDescription"]
    )["itemDescription"]
    .count()
    .unstack()
    .fillna(0))
```

```python
basket = basket.applymap(
    lambda x: 1 if x > 0 else 0)
```

```python
basket.shape
```

```python
basket.head()
```

```python
top_products.head(10)
```

```python
basket.shape
```

```python
!pip install mlxtend -q
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(
    basket,
    min_support=0.01,
    use_colnames=True)

frequent_itemsets.sort_values(
    by="support",
    ascending=False).head(20)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1)

rules.head()
```

```python
rules.sort_values(by="lift",
    ascending=False).head(20)
```

```python
strong_rules = rules[
    (rules["confidence"] > 0.30) &
    (rules["lift"] > 1.2)]

strong_rules.sort_values(
    by="lift",
    ascending=False).head(20)
```

```python
print(rules.shape)
```

```python
frequent_itemsets = apriori(
    basket,
    min_support=0.001,
    use_colnames=True)

print(frequent_itemsets.shape)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.05)

print(rules.shape)
```

```python
rules[[ "antecedents",
        "consequents",
        "support",
        "confidence",
        "lift" ]].sort_values(by="lift",ascending=False).head(20)
```

## Association Rule Analysis

The Apriori algorithm successfully identified product associations within customer transactions.

One of the strongest rules discovered was:

* Yogurt and Whole Milk → Sausage

This rule achieved a lift value of 2.18, indicating that customers who purchased yogurt and whole milk were more than twice as likely to purchase sausage compared to the average customer.

Other meaningful associations included:

* Whole Milk and Sausage → Yogurt
* Specialty Chocolate → Citrus Fruit
* Whole Milk and Soda → Sausage

These findings demonstrate how market basket analysis can help retailers optimize product placement, create bundle promotions, and improve cross-selling strategies.
